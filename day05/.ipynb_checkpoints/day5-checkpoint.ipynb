```python
def makeStacks(data):
    stacks = []
    for d in data:
        if d == '':
            return stacks
        else:
            stacks += [[d[x:x+3] for x in range(0,len(d),4)]]
```

```python
def makeS(stacks):
    s=[[] for _ in range(int(stacks[-1][-1]))]
    for x in range(len(stacks)-1,-1,-1):
        for i,obj in enumerate(stacks[x]):
            if obj == '   ':
                pass
            else:
                s[i].append(obj)
    return(s)
```

```python
def buildStacks(data):
    return makeS(makeStacks(data))
```

```python
stacks = buildStacks(data)
```

```python
stacks
```

```python
def getInstructions(data):
    inst = []
    instructions = False
    for d in data:
        if instructions:
            #inst += [[x for x in d.split(' ') if len(x) != 3]]
            inst += [d.split(' ')[1::2]]
        if d == '':
            instructions = True
    return inst
```

```python
instr = getInstructions(data)
```

```python
def carryOutInstructions(stack, move, frm, to):
    for x in range(int(move)):
        stack[int(to)-1].append(stack[int(frm)-1].pop())
    return stack
```

```python
instr
```

```python
for x in instr:
    stacks = carryOutInstructions(stacks,x[0],x[1],x[2])
print(stacks)
```

```python
out = ''
for x in stacks:
    out += x[-1][1]
print(out)
```

```python
# part one Toy example
in_file = "toy.txt"
with open(in_file) as f:
    lines = f.readlines()
data = [l.strip('\n') for l in lines]

stacks = buildStacks(data)
instr = getInstructions(data)
for x in instr:
    stacks = carryOutInstructions(stacks,x[0],x[1],x[2])
out = ''
for x in stacks:
    out += x[-1][1]
print(out)
```
