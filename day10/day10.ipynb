```python
def read(command):
    try:
        cmd, val = command.split(' ')
        return cmd, int(val)
    except:
        return command, 0
```

```python
class Instruction:
    def __init__(self, tick, val):
        self.tick = int(tick)
        self.val = int(val)
```

```python
def addx(tick, val):
    return Instruction(tick+1, val)
```

```python
signal_strength = 0
X = 1
to_do = True
tick = 1
queue = []
while to_do:
    try:
        cmd, val = read(data[tick -1])
    except:
        #print("no more instructions just work on queue")
        cmd, val = "run_on",0
    try:
        next_free = queue[-1].tick + 1
    except:
        #print("run once")
        next_free = tick
    if cmd == 'addx':
        queue.append(addx(next_free,val))
    elif cmd == 'run_on':
        pass
    else:
        queue.append(Instruction(next_free,0))
    try:
        #print(tick, queue[0].tick)
        if queue[0].tick == tick:
            x = queue.pop(0)
            X += x.val
        else:
            pass
    except:
        to_do = False
    tick += 1
    if (tick - 20) %  40 == 0:
        print(tick, X, "=", tick*X)
        signal_strength += tick * X
    print("after ",tick,"X=",X)
print(X)
print(signal_strength)
```
