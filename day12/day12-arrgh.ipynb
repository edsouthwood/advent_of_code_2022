```python
# returns valid moves and bool = to hill climb True or False

def up(data, x, y):
    if y-1 < 0:
        return None
    if data[y][x] == 'z':
        print('z')
        if (data[y-1][x] == 'E'):
            return (-1,-1)
    else:
        if ord(data[y][x]) == ord(data[y-1][x]):
            return (x,y-1)
        elif ord(data[y][x]) + 1 == ord(data[y-1][x]):
            return (x,y-1)
        else:
            return None
def down(data, x, y):
    try:
        if data[y][x] == 'z':
            if data[y+1][x] == 'E':
                return (-1,-1)
        if ord(data[y][x]) == ord(data[y+1][x]):
            return (x,y+1)
        elif ord(data[y][x]) + 1 == ord(data[y+1][x]):
            return (x,y+1)
        else:
            return None
    except IndexError:
        return None
def left(data, x, y):
    if x-1 < 0:
        return None
    if data[y][x] == 'z':
        if data[y][x-1] == 'E':
            return (-1,-1)
    else:
        if ord(data[y][x]) == ord(data[y][x-1]):
            return (x-1,y)
        elif ord(data[y][x]) + 1 == ord(data[y][x-1]):
            return (x-1,y)
        else:
            return None
def right(data, x, y):
    try:
        if data[y][x] == 'z':
            if data[y][x+1] == 'E':
                return (-1,-1)
        if ord(data[y][x]) == ord(data[y][x+1]):
            return (x+1,y)
        elif ord(data[y][x]) + 1 == ord(data[y][x+1]):
            return (x+1,y)
        else:
            return None
    except IndexError:
        return None
```

```python
def tryNeighbours(point, data):
    possible_points = []
    possible_points += [up(data, point[0], point[1])]
    possible_points += [down(data, point[0], point[1])]
    possible_points += [left(data, point[0], point[1])]
    possible_points += [right(data, point[0], point[1])]
    return [i for i in possible_points if i is not None]
```

```python
class distanceMap:
    def __init__(self,data):
        rows = len(data)
        columns = len(data[0])
        distanceMap = []
        for r in data:
            distanceMap += [[999]* columns]
        self.distanceMap = distanceMap
    
    def setStart(self, point):
        self.distanceMap[point[1]][point[0]] = 0
    
    def update(self, from_point, to_point):
        #print(from_point,to_point)
        from_point_val = self.distanceMap[from_point[1]][from_point[0]]
        to_point_val = self.distanceMap[to_point[1]][to_point[0]]
        #print(from_point_val,to_point_val)
        if (from_point_val + 1) < to_point_val:
            self.distanceMap[to_point[1]][to_point[0]] = from_point_val + 1
```

```python
dmap = distanceMap(data)
start = (0,20)
dmap.setStart(start)
queue = []
queue.append(start)
visited = set()
while queue:
    #print(queue)
    head = queue.pop(0)
    visited.add(head)
    possible_points = tryNeighbours(head, data)
    for p in possible_points:
        dmap.update(head,p)
        if p in visited:
            pass
        else:
            queue.append(p)
```
