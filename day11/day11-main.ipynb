```python
in_file = "input.txt"
with open(in_file) as f:
    lines = f.readlines()
data = [l.strip('\n') for l in lines]

# define monkey

class Monkey:
    def __init__(self, items, opp, test, tr, fa):
        items = [int(i.lstrip()) for i in items.split(',')]
        self.items = items
        self.opp = opp
        self.test = int(test)
        self.tr = int(tr)
        self.fa = int(fa)
        self.inspections = 0
    
    def __repr__(self) -> str:
        return f'Monkey("{self.items}","{self.opp}","{self.test}","{self.tr}","{self.fa}")'
    
    def addItem(self, new_item):
        return self.items.append(new_item)
    
    def inspect(self):
        self.inspections += len(self.items) 
        _, cmd, val = self.opp.split(' ')
        if cmd == "+":
            for i, v in enumerate(self.items):
                v = v+int(val)
                v = v // 3
                self.items[i] = v
        if cmd == "*":
            if val == "old":
                for i, v in enumerate(self.items):
                    v = v*v
                    v = v // 3
                    self.items[i] = v
            else:
                for i, v in enumerate(self.items):
                    v = v*int(val)
                    v = v // 3
                    self.items[i] = v
    
    def testItems(self):
        vals = []
        targets = []
        for i, v in enumerate(self.items):
            vals += [v] 
            if (v % self.test) == 0:
                targets += [self.tr]
            else:
                targets += [self.fa]
        self.items = []
        return list(zip(targets,vals))

# parse input

items = []
monkeys = []
index = 0
for d in data:
    if d == "":
        index += 1
        #print("next monkey")
        #print(tr)
        #print(fa)
        #print(items)
        monkeys.append(Monkey(items,opp,test,tr,fa))
        items = []
    else:
        d = d.lstrip()
        if d[0] == "M":
            pass
            #print(d)
        elif d[0] == "S":
            """items_to_sort = d.split(' ')
            for i in items_to_sort:
                try:
                    i = int(i)
                    items += [i]
                except:
                    pass"""
            d = d.split(':')
            d = d[1].lstrip()
            items = d
            #print(d, items)
        elif d[0] == "O":
            opp = d.split('=')
            opp = opp[-1].lstrip()
            #print(d, opp)
        elif d[0] == "T":
            test = d.split('by')
            test = test[-1].lstrip()
            #print(d, test)
        elif d[0] == "I":
            d = d.split(' ')
            if d[1] == 'true:':
                tr = d[-1]
                #print(d[1], tr)
            else:
                fa = d[-1]
                #print(d[1], fa)

# perform monkey business

for _ in range(20):
    for m in monkeys:
        m.inspect()
        thrown_items = m.testItems()
        for i in thrown_items:
            monkeys[i[0]].addItem(i[1])
business = []
for m in monkeys:
    print(m.inspections)
    business += [m.inspections]
business.sort()
print(business[-1]*business[-2])
```
